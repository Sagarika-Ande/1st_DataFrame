```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# deleting and cleaning the null values from dataset
data_clean=data.dropna()
print(data_clean)
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.countplot(x='ClientName', hue='State', data=data_clean)
plt.title('Business', fontsize=12)
plt.xlabel('Client Name', fontsize=10)
plt.ylabel('Count', fontsize=10)
plt.xticks(rotation=45, ha='right', fontsize=8)  # Rotate x-axis labels if needed
plt.legend(title='State', fontsize=8)
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```

```python
top_enterprises = data_clean['EnterpriseCode'].value_counts().dropna().head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_enterprises.values, y=top_enterprises.index, palette="Blues_d")
plt.title("Top 10 EnterpriseCodes by User Count")
plt.xlabel("User Count")
plt.ylabel("EnterpriseCode")
plt.tight_layout()
plt.show()
```

```python

# Group and count how many users are tied to each (EnterpriseCode, ClientName) pair
grouped = data_clean.groupby(['EnterpriseCode', 'ClientName']).size().reset_index(name='UserCount')

# Drop missing ClientNames for clarity
grouped = grouped.dropna(subset=['ClientName'])

# Sort to get most active pairs
top_pairs = grouped.sort_values(by='UserCount', ascending=False).head(20)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_pairs, x='UserCount', y='ClientName', hue='EnterpriseCode', dodge=False)
plt.title("Top EnterpriseCode and ClientName Combinations")
plt.xlabel("User Count")
plt.ylabel("Client Name")
plt.legend(title="EnterpriseCode", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
zip_counts = data_clean['Zipcode'].value_counts().dropna().head(10)  # Top 10 Zipcodes

# Create a DataFrame with Zipcode and corresponding City
zip_city_df = data_clean[['Zipcode', 'City']].drop_duplicates()
zip_city_df = zip_city_df[zip_city_df['Zipcode'].isin(zip_counts.index)]

# Merge zip_counts with zip_city_df to get City information
zip_counts_with_city = pd.merge(zip_counts.rename('Count'), zip_city_df, on='Zipcode')

fig, ax1 = plt.subplots(figsize=(10, 6))
sns.barplot(x=zip_counts_with_city['Count'], y=zip_counts_with_city['Zipcode'].astype(str), palette='magma', ax=ax1)
ax1.set_title("Top 10 Zipcodes by User Count")
ax1.set_xlabel("User Count")
ax1.set_ylabel("Zipcode")

ax2 = ax1.twinx()
# Use the merged DataFrame for the line plot
sns.lineplot(x=zip_counts_with_city['Count'], y=zip_counts_with_city['City'], marker='o', color='red', ax=ax2)
ax2.set_ylabel("City")

plt.show()
```

```python
# Convert 'trendDate' to datetime format
data_clean['trendDate'] = pd.to_datetime(data_clean['trendDate'], errors='coerce')

# Drop rows with invalid or missing dates
df = data_clean.dropna(subset=['trendDate'])

# Extract 'Year-Month' for grouping
data_clean['YearMonth'] = data_clean['trendDate'].dt.to_period('M')

# Group by Year-Month and count users
monthly_activity = data_clean.groupby('YearMonth').size()

# Plot the trend
plt.figure(figsize=(12, 6))
monthly_activity.plot(kind='line', marker='o', color='teal')
plt.title('User Activity Over Time (Monthly)')
plt.xlabel('Year-Month')
plt.ylabel('Number of Users')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
